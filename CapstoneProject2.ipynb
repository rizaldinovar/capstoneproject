## 1.Tangani missing values, error, duplikasi, dan outlier di setiap kolom

```python
#Konversi kolom numerik dari object (jika belum)
numerical_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Cek missing values awal
print("Jumlah missing values sebelum imputasi:")
display(df.isnull().sum())

# Tangani missing values:
# - numerik => isi dengan rata-rata (mean)
# - kategorikal => isi dengan modus (mode)
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Cek hasil
print("\nJumlah missing values setelah imputasi:")
display(df.isnull().sum())
```

```python
# Cek jumlah data duplikat
print(f"Jumlah data sebelum menghapus duplikat: {df.shape[0]}")
duplicates = df.duplicated().sum()
print(f"Jumlah data duplikat yang ditemukan: {duplicates}")

# Hapus duplikat
df.drop_duplicates(inplace=True)

# Konfirmasi
print(f"Jumlah data setelah menghapus duplikat: {df.shape[0]}")
```

```python
# Hitung IQR untuk setiap kolom numerik
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Tentukan data yang berada dalam batas wajar
condition = ~((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)

# Tampilkan jumlah data sebelum dan sesudah
print(f"Jumlah data sebelum menghapus outlier: {df.shape[0]}")
df = df[condition]
print(f"Jumlah data setelah menghapus outlier: {df.shape[0]}")
```

## 2.Ubah data kategori menjadi numerik

```python
from sklearn.preprocessing import LabelEncoder

# Daftar kolom kategorikal (selain target)
categorical_cols = [
    'Gender', 'CALC', 'FAVC', 'SCC', 'SMOKE',
    'family_history_with_overweight', 'CAEC', 'MTRANS'
]

# Pastikan semua kolom sudah tidak ada missing values
for col in categorical_cols:
    df[col] = df[col].astype(str)  # pastikan bertipe string

# Lakukan one-hot encoding dan hilangkan kolom pertama (drop_first=True) untuk menghindari dummy trap
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
# Label encoding target
le = LabelEncoder()
df_encoded['NObeyesdad'] = le.fit_transform(df_encoded['NObeyesdad'])

# Simpan nama-nama kelas jika perlu ditampilkan nanti
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label Encoding untuk Target (NObeyesdad):")
print(label_mapping)
```

```python
# Pisahkan fitur dan target
X = df_encoded.drop('NObeyesdad', axis=1)
y = df_encoded['NObeyesdad']
```

## 3.Tentukan apakah semua feature perlu digunakan

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gabungkan kembali X dan y untuk analisis korelasi
df_corr = X.copy()
df_corr['NObeyesdad'] = y

# Hitung korelasi Pearson
plt.figure(figsize=(14, 12))
corr_matrix = df_corr.corr()

# Tampilkan korelasi fitur dengan target
sns.heatmap(corr_matrix[['NObeyesdad']].sort_values(by='NObeyesdad', ascending=False),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korelasi Fitur terhadap Target (NObeyesdad)")
plt.show()
```

## 4.Atasi ketidakseimbangan kelas data

```python
# Cek distribusi kelas target
import matplotlib.pyplot as plt

print("Distribusi Kelas Sebelum SMOTE:")
print(y.value_counts())

# Visualisasi distribusi kelas
plt.figure(figsize=(8,5))
y.value_counts().plot(kind='bar')
plt.title("Distribusi Kelas Sebelum Penyeimbangan")
plt.xlabel("Label (NObeyesdad)")
plt.ylabel("Jumlah")
plt.xticks(rotation=0)
plt.show()
```

```python
# Gunakan SMOTE untuk oversampling kelas minoritas
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Cek distribusi setelah SMOTE
print("Distribusi Kelas Setelah SMOTE:")
print(pd.Series(y_resampled).value_counts())

# Visualisasi setelah SMOTE
plt.figure(figsize=(8,5))
pd.Series(y_resampled).value_counts().plot(kind='bar', color='orange')
plt.title("Distribusi Kelas Setelah Penyeimbangan (SMOTE)")
plt.xlabel("Label (NObeyesdad)")
plt.ylabel("Jumlah")
plt.xticks(rotation=0)
plt.show()
```
