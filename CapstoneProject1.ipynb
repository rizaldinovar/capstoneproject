```python
# Tampilkan beberapa baris pertama dan informasi dataset
import pandas as pd

# Load dataset
file_path = '/content/drive/My Drive/Colab Notebooks/CapstoneBengkelKoding/ObesityDataSet.csv'
df = pd.read_csv(file_path)

# Tampilkan beberapa baris pertama
print("Beberapa baris pertama:")
display(df.head())

# Jumlah baris dan kolom
print(f"\nJumlah baris: {df.shape[0]}")
print(f"Jumlah kolom: {df.shape[1]}")

# Tipe data sebelum diperbaiki
print("\nTipe data sebelum konversi:")
print(df.dtypes)

# Konversi kolom numerik yang salah tipe (terbaca sebagai object)
columns_to_convert = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Cek ulang tipe data
print("\nTipe data setelah konversi:")
print(df.dtypes)

# Statistik deskriptif (hanya kolom numerik)
print("\nStatistik deskriptif kolom numerik:")
display(df.describe())
```

## 2.Tampilkan visualisasi data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ambil kolom numerik yang valid
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Visualisasi distribusi fitur numerik
if len(numerical_cols) > 0:
    df[numerical_cols].hist(bins=30, figsize=(15, 10), color='skyblue')
    plt.suptitle("Distribusi Fitur Numerik")
    plt.show()

    # Korelasi antar fitur numerik
    plt.figure(figsize=(10, 8))
    sns.heatmap(df[numerical_cols].corr(), annot=True, cmap="coolwarm")
    plt.title("Heatmap Korelasi Fitur Numerik")
    plt.show()
else:
    print("Tidak ditemukan kolom numerik valid.")
```

## 3.Cek keberadaan missing values, unique value, data duplikat, keseimbangan data, dan deteksi outlier menggunakan boxplot

```python
print("Jumlah Missing Values di setiap kolom:")
missing = df.isnull().sum()
display(missing[missing > 0] if missing.sum() > 0 else "Tidak ada missing values.")
```

```python
print("Jumlah Nilai Unik per Kolom:")
unique_vals = df.nunique().sort_values()
display(unique_vals)
```

```python
print("Jumlah Data Duplikat:")
duplicate_count = df.duplicated().sum()
print(f"Terdapat {duplicate_count} data duplikat.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

print("Distribusi Kelas Target (NObeyesdad):")
display(df['NObeyesdad'].value_counts())

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='NObeyesdad', order=df['NObeyesdad'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Distribusi Kategori Obesitas (Target)")
plt.xlabel("Kategori Obesitas")
plt.ylabel("Jumlah")
plt.tight_layout()
plt.show()
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Deteksi Outlier Menggunakan Boxplot:")
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot: {col}')
    plt.tight_layout()
    plt.show()
```
