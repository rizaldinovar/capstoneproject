## 1.Tangani missing values, error, duplikasi, dan outlier di setiap kolom

```python
#Konversi kolom numerik dari object (jika belum)
numerical_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Cek missing values awal
print("Jumlah missing values sebelum imputasi:")
display(df.isnull().sum())

# Tangani missing values:
# - numerik => isi dengan rata-rata (mean)
# - kategorikal => isi dengan modus (mode)
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Cek hasil
print("\nJumlah missing values setelah imputasi:")
display(df.isnull().sum())
```

```python
# Cek jumlah data duplikat
print(f"Jumlah data sebelum menghapus duplikat: {df.shape[0]}")
duplicates = df.duplicated().sum()
print(f"Jumlah data duplikat yang ditemukan: {duplicates}")

# Hapus duplikat
df.drop_duplicates(inplace=True)

# Konfirmasi
print(f"Jumlah data setelah menghapus duplikat: {df.shape[0]}")
```

```python
# Hitung IQR untuk setiap kolom numerik
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Tentukan data yang berada dalam batas wajar
condition = ~((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)

# Tampilkan jumlah data sebelum dan sesudah
print(f"Jumlah data sebelum menghapus outlier: {df.shape[0]}")
df = df[condition]
print(f"Jumlah data setelah menghapus outlier: {df.shape[0]}")
```

## 2.Ubah data kategori menjadi numerik

```python
from sklearn.preprocessing import LabelEncoder

# Daftar kolom kategorikal (selain target)
categorical_cols = [
    'Gender', 'CALC', 'FAVC', 'SCC', 'SMOKE',
    'family_history_with_overweight', 'CAEC', 'MTRANS'
]

# Pastikan semua kolom sudah tidak ada missing values
for col in categorical_cols:
    df[col] = df[col].astype(str)  # pastikan bertipe string

# Lakukan one-hot encoding dan hilangkan kolom pertama (drop_first=True) untuk menghindari dummy trap
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
# Label encoding target
le = LabelEncoder()
df_encoded['NObeyesdad'] = le.fit_transform(df_encoded['NObeyesdad'])

# Simpan nama-nama kelas jika perlu ditampilkan nanti
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label Encoding untuk Target (NObeyesdad):")
print(label_mapping)
```

```python
# Pisahkan fitur dan target
X = df_encoded.drop('NObeyesdad', axis=1)
y = df_encoded['NObeyesdad']
```

## 3.Lakukan evaluasi dengan menggunakan confusion matrix, akurasi, presisi, recall, dan F1- score

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, ConfusionMatrixDisplay

# Fungsi evaluasi model
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted')  # weighted untuk multi-kelas
    rec = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    print(f"--- Evaluasi {model_name} ---")
    print(f"Akurasi : {acc:.4f}")
    print(f"Presisi : {prec:.4f}")
    print(f"Recall  : {rec:.4f}")
    print(f"F1-Score: {f1:.4f}\n")

    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()

    return acc, prec, rec, f1

# Evaluasi ketiga model
evaluate_model(log_reg, X_test, y_test, "Logistic Regression")
evaluate_model(rf_clf, X_test, y_test, "Random Forest")
evaluate_model(gb_clf, X_test, y_test, "Gradient Boosting")
```

## 1.Lakukan optimasi hyperparameter tuning menggunakan GridSearchCV atau RandomizedSearchCV

```python
# Import library yang dibutuhkan
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# --- Logistic Regression: Gunakan GridSearchCV
logreg_param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs']
}

logreg_grid = GridSearchCV(
    estimator=LogisticRegression(max_iter=1000),
    param_grid=logreg_param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
logreg_grid.fit(X_train, y_train)

# --- Random Forest: Gunakan RandomizedSearchCV
rf_param_dist = {
    'n_estimators': [50, 100, 150, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

rf_random = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=rf_param_dist,
    n_iter=10,
    cv=5,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)
rf_random.fit(X_train, y_train)

# --- Gradient Boosting: Gunakan GridSearchCV
gb_param_grid = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

gb_grid = GridSearchCV(
    estimator=GradientBoostingClassifier(random_state=42),
    param_grid=gb_param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
gb_grid.fit(X_train, y_train)

# Tampilkan parameter terbaik untuk masing-masing model
print("Best Parameters Logistic Regression:", logreg_grid.best_params_)
print("Best Parameters Random Forest:", rf_random.best_params_)
print("Best Parameters Gradient Boosting:", gb_grid.best_params_)
```

## 2.Latih ulang model dengan parameter terbaik dan evaluasi kembali untuk melihat peningkatan performa.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score

# Fungsi evaluasi model dengan visualisasi confusion matrix
def evaluate_model_with_plot(name, model, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='weighted')
    rec = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')

    print(f"=== {name} ===")
    print(f"Akurasi : {acc:.4f}")
    print(f"Presisi : {prec:.4f}")
    print(f"Recall  : {rec:.4f}")
    print(f"F1-score: {f1:.4f}")
    print("Classification Report:\n", classification_report(y_true, y_pred))

    # Confusion matrix visual
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.show()
    print("\n")

# Latih ulang model terbaik
log_reg_best = LogisticRegression(C=10, solver='lbfgs', max_iter=1000, random_state=42)
rf_best = RandomForestClassifier(n_estimators=200, min_samples_split=5, max_depth=30, random_state=42)
gb_best = GradientBoostingClassifier(learning_rate=0.1, max_depth=7, n_estimators=100, random_state=42)

log_reg_best.fit(X_train, y_train)
rf_best.fit(X_train, y_train)
gb_best.fit(X_train, y_train)

# Prediksi
y_pred_log_best = log_reg_best.predict(X_test)
y_pred_rf_best = rf_best.predict(X_test)
y_pred_gb_best = gb_best.predict(X_test)

# Evaluasi dengan visualisasi
evaluate_model_with_plot("Logistic Regression (Tuned)", log_reg_best, y_test, y_pred_log_best)
evaluate_model_with_plot("Random Forest (Tuned)", rf_best, y_test, y_pred_rf_best)
evaluate_model_with_plot("Gradient Boosting (Tuned)", gb_best, y_test, y_pred_gb_best)
```

## 3.Tampilkan visualisasi perbandingan performa antar model sebelum dan sesudah optimasi.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# --- Nilai metrik SEBELUM tuning (sesuai data kamu)
before_tuning = {
    'Model': ['Logistic Regression', 'Random Forest', 'Gradient Boosting'],
    'Accuracy': [0.8277, 0.9476, 0.9588],
    'Precision': [0.8256, 0.9536, 0.9603],
    'Recall': [0.8277, 0.9476, 0.9588],
    'F1-Score': [0.8249, 0.9481, 0.9585]
}

# --- Nilai metrik SESUDAH tuning (ambil dari evaluasi model terbaik yang sudah dihitung)
after_tuning = {
    'Model': ['Logistic Regression', 'Random Forest', 'Gradient Boosting'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_log_best),
        accuracy_score(y_test, y_pred_rf_best),
        accuracy_score(y_test, y_pred_gb_best)
    ],
    'Precision': [
        precision_score(y_test, y_pred_log_best, average='weighted'),
        precision_score(y_test, y_pred_rf_best, average='weighted'),
        precision_score(y_test, y_pred_gb_best, average='weighted')
    ],
    'Recall': [
        recall_score(y_test, y_pred_log_best, average='weighted'),
        recall_score(y_test, y_pred_rf_best, average='weighted'),
        recall_score(y_test, y_pred_gb_best, average='weighted')
    ],
    'F1-Score': [
        f1_score(y_test, y_pred_log_best, average='weighted'),
        f1_score(y_test, y_pred_rf_best, average='weighted'),
        f1_score(y_test, y_pred_gb_best, average='weighted')
    ]
}

# Convert to DataFrames
df_before = pd.DataFrame(before_tuning)
df_after = pd.DataFrame(after_tuning)

# Plotting
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
x = np.arange(len(df_before['Model']))  # posisi model
width = 0.35  # lebar bar

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

for i, metric in enumerate(metrics):
    axs[i].bar(x - width/2, df_before[metric], width, label='Before Tuning', color='skyblue')
    axs[i].bar(x + width/2, df_after[metric], width, label='After Tuning', color='orange')
    axs[i].set_title(f'{metric} Comparison')
    axs[i].set_xticks(x)
    axs[i].set_xticklabels(df_before['Model'], rotation=15)
    axs[i].set_ylim(0, 1)
    axs[i].legend()
    axs[i].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.suptitle('Perbandingan Performa Model Sebelum & Sesudah Hyperparameter Tuning', fontsize=16, y=1.02)
plt.show()
```
